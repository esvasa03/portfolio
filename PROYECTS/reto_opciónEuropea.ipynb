Ahora, para las opciones put, el comprador tiene derecho a vender la acción a un precio y tiempo determinado, en este caso, el vendedor es el que tiene ese derecho, por lo tanto debe pagar su respectiva prima, por el otro lado, el comprador está obligado a comprar la acción.</h3>

        <h3><i>2.2. Movimiento Browniano Geométrico</i></h3>
        <h3>&emsp;El movimiento Browniano geométrico (GBM, por sus siglas en inglés) es un proceso estocástico que se utiliza para modelar el comportamiento de los precios de los activos financieros en el tiempo. Es una extensión del movimiento Browniano, que fue propuesto por el botánico Robert Brown en 1827 y posteriormente formalizado matemáticamente por Albert Einstein en 1905 (Mosiño et al., 2019).</h3>
        <h3>&emsp;El GBM se diferencia del movimiento Browniano estándar en que tiene una tendencia de crecimiento exponencial, lo que lo hace más adecuado para modelar precios de activos que no pueden ser negativos. La ecuación diferencial estocástica que define el GBM es:</h3>
        <h3>\begin{equation}
        dS_t = \mu S_t dt + \sigma S_t dW_t
        \end{equation}</h3>
        <h3>&emsp;Donde:</h3>
        <ul>
          <li>\(S_t\) es el precio del activo en el tiempo \(t\).</li>
          <li>\(\mu\) es la tasa de retorno esperada del activo.</li>
          <li>\(\sigma\) es la volatilidad del activo.</li>
          <li>\(dW_t\) es un término de movimiento Browniano, que representa la variación aleatoria del precio del activo.</li>
        </ul>
        <h3>&emsp;Una característica importante del GBM es que los rendimientos logarítmicos del precio del activo, definidos como $\ln\left(\frac{S_t}{S_{t-1}}\right)$, son normalmente distribuidos, lo que facilita el cálculo de probabilidades y la valoración de opciones y otros derivados financieros.</h3>
    </div>
</font>

# *2.3 Modelo de Black-Scholes*

El modelo de Black-Scholes, desarrollado por Fischer Black y Myron Scholes en 1973, es un modelo matemático que se utiliza para calcular el precio teórico de opciones europeas y otros derivados financieros. La fórmula de Black-Scholes es una de las herramientas más importantes y fundamentales en el campo de las finanzas cuantitativas y ha sido ampliamente utilizada para valorar opciones (Benito, 2012).

La fórmula de Black-Scholes para una opción de compra (call) europea es:

$
C = S_0 e^{-qT} N(d_1) - X e^{-rT} N(d_2)
$

Y para una opción de venta (put) europea es:

$
P = X e^{-rT} N(-d_2) - S_0 e^{-qT} N(-d_1)
$

Donde:
- $(C$) y $(P$) son los precios de la opción de compra y de venta, respectivamente.
- $(S_0$) es el precio actual del activo subyacente.
- $(X$) es el precio de ejercicio de la opción.
- $(T$) es el tiempo hasta el vencimiento de la opción.
- $(r$) es la tasa libre de riesgo.
- $(q$) es la tasa de dividendos del activo subyacente.
- $(N(d_1)$) y $(N(d_2)$) son los valores de la función de distribución acumulativa de una distribución normal estándar para $(d_1$) y $(d_2$), respectivamente.
- $(d_1$) y $(d_2$) se calculan como:

$
d_1 = \frac{\ln\left(\frac{S_0}{X}\right) + \left(r - q + \frac{\sigma^2}{2}\right)T}{\sigma\sqrt{T}}
$

$
d_2 = d_1 - \sigma\sqrt{T}
$

- (σ) es la volatilidad del activo subyacente.

Este modelo solo funciona bajo los siguientes supuestos:
1. El comportamiento del precio de la acción corresponde al modelo logarítmico normal, con u y σ constantes.
2. No hay costos de transacción ni impuestos. Todos los títulos son perfectamente divisibles.
3. No hay dividendos sobre la acción durante la vida de la opción.
4. No hay oportunidades de arbitraje libres de riesgo.
5. La negociación de valores es continua.
6. Los inversionistas pueden adquirir u otorgar préstamos a la misma tasa de interés libre de riesgo.

7. La tasa de interés libre de riesgo a corto plazo, $r$, es constante

Una de las mayores contribuciones del modelo de Black-Scholes fue proporcionar una fórmula cerrada para el precio de las opciones, lo que simplificó significativamente el proceso de valoración de las mismas. El modelo también ayudó a popularizar el concepto de "volatilidad implícita", que es la volatilidad del activo subyacente que hace que el precio de mercado de la opción sea igual al precio teórico calculado por la fórmula de Black-Scholes.

        <h3><i>2.4 Método Arch de cadena de Markov </i></h3>
        <h3>&emsp;El Método Arch (Autoregressive Conditional Heteroskedasticity) de cadena de Markov es una técnica estadística que se utiliza para modelar y predecir la volatilidad de las series temporales financieras. Este método combina la modelización de la heterocedasticidad condicional autoregresiva (ARCH) con un proceso de cadena de Markov, permitiendo capturar la dependencia temporal de la volatilidad y las transiciones entre diferentes estados o regímenes de volatilidad (Steel, 2008).</h3>
        <h3>&emsp;El modelo ARCH fue introducido por Robert Engle en 1982 como una forma de modelar la volatilidad condicional heterocedástica en series temporales financieras. Posteriormente, Tim Bollerslev generalizó el modelo ARCH a un modelo GARCH (Generalized ARCH), que incluye términos de varianza pasada para mejorar la predicción de la volatilidad futura (Bildirici & Ersin, 2014).</h3>
        <h3>&emsp;La integración de la cadena de Markov en el método ARCH permite representar el proceso estocástico de la volatilidad como un sistema que transita entre diferentes estados de volatilidad, cada uno con su propia dinámica y parámetros de modelo. Esto permite capturar la naturaleza no lineal y no estacionaria de la volatilidad financiera, ofreciendo una representación más completa y precisa de la misma.</h3>

        <h3><i>2.5 Método de montecarlo</i></h3>
        <h3>&emsp;El método Monte Carlo es una poderosa técnica computacional para aproximar soluciones a problemas mediante muestreo aleatorio. Una vez que tenemos un modelo del comportamiento del activo, podemos generar una gran cantidad de simulaciones de los precios futuros del activo tomando muestras repetidas del modelo y calculando los precios futuros en función de los rendimientos logarítmicos muestreados.</h3>
        <h3>&emsp;En el área de las finanzas, los precios financieros deberían mostrar un patrón de caminata aleatoria. Más precisamente, se supone que los precios siguen un proceso de Markov , que es un proceso estocástico particular donde toda la distribución del precio futuro depende únicamente del precio actual (Polanitzer, 2023).</h3>
        <h3>&emsp;El método de montecarlo se obtiene por medio de la simulación de una caminata aleatoria por medio de la siguiente fórmula:</h3>
        <h3>\begin{equation}
        S_{T} = S_{0} e^{\mu T}
        \end{equation}</h3>
        <h3>&emsp;Donde $S_0$ y $S_T$ son el precio de las acciones en el momento cero y en el momento T . La ecuación anterior muestra que, cuando la tasa de variación es cero, el precio de las acciones crece a una tasa compuesta continuamente de μ por unidad de tiempo. El parámetro μ es la tasa de rendimiento esperada de la acción.</h3>

```python
# Movimiento Browniano Geométrico
def gbm(nsteps = 10000, t = 1, mu = 0.001, sigma = 0.02, start = 1):
  steps = [(mu - (sigma**2)/2)*t + np.random.randn()*sigma for i in range(nsteps)]
  y = start*np.exp(np.cumsum(steps))
  x = [t*i for i in range(nsteps)]
  return x, y
```

### Black Scholes

```python
class BlackScholes:

    def __init__(self,S,K,T,r,sigma):
        self.S = S
        self.K = K
        self.T = T
        self.r = r
        self.sigma = sigma

    def d1(self):
        return(log(self.S/self.K)+(self.r+self.sigma**2*0.5)*self.T)/(self.sigma*sqrt(self.T))

    def d2(self):
        return self.d1()-self.sigma*sqrt(self.T)

    def call(self):
        return self.S*norm.cdf(self.d1())-self.K*exp(-self.r*self.T)*norm.cdf(self.d2())

    def put(self):
        return self.K * exp(-self.r * self.T) * norm.cdf(-self.d2()) - self.S * norm.cdf(-self.d1())

    def call_delta(self):
      return norm.cdf(self.d1())

    def put_delta(self):
      return -norm.cdf(-self.d1())

    def gamma(self):
      return norm.pdf(self.d1())/(self.S*self.sigma*sqrt(self.T))

#(Saqan, 2020)
```

```python
returns = np.log(stock['Adj Close']).diff().dropna()

sigma = np.sqrt(252) * returns.std()

expire = end + timedelta(days=30)
t = (datetime.strptime(expire.strftime("%m-%d-%Y"), "%m-%d-%Y") - datetime.utcnow()).days / 365

So = stock['Close'].iloc[-1]

ticker_symbol = "^TNX"
data = yf.download(ticker_symbol, start=start, end=end)
r = data['Adj Close'].iloc[-1]/100
```

```python
sp1 = So + (So*0.1)
bs_put = BlackScholes(S=So, K=sp1, T=t, r=r, sigma=sigma)
print('El valor de la opción pull es: ', bs_put.put())
```

En esta sección, se ha realizado un análisis de sensibilidad para entender cómo el precio de las opciones call y put cambian con respecto al precio del activo subyacente y la volatilidad. Estos análisis son útiles para comprender el riesgo asociado con las fluctuaciones en estos parámetros y para tomar decisiones informadas al operar con opciones.

Primero, se puede entender el comportamiento que tienen el precio de las opciones call y put respecto al precio de ejercicio. Lo que podemos analizar es que el precio de las opciones call y de las acciones tiene un comportamiento inversamente proporcional, donde mientras más alto sea el precio de ejercicio, menor es el precio de la opción. Este fenómeno se invierte con el precio de las opciones put, donde entre menor sea el precio de ejercicio, menor será el precio de la opción.

Por otro lado, lo que podemos observar cuando modificamos los valores de r (tipo de interés libre de riesgo) y T (tiempo que le resta de vida a la opción), es que el precio de los dos tipos de opciones tiende a incrementar. Por medio de esta suposición podemos llegar a la conclusión de que se debe al riesgo que puede tener la inversión. Debido a que un mayor interés libre de riesgo, significa que el activo puede tener un mayor riesgo; así como, un mayor periodo, pero conllevar una mayor incertidumbre en el comportamiento de una acción.

Los números griegos son medidas importantes en la gestión de riesgos y en la toma de decisiones financieras, especialmente cuando se opera con opciones.

Delta mide la sensibilidad del precio de una opción respecto a cambios en el precio del activo subyacente. La delta de una opción varía entre 0 y 1 para call; y -1 a 0 para put. El delta indica cuánto cambiará el precio de la opción en respuesta a un cambio unitario en el precio del activo subyacente. Así mismo, la delta de una opción también puede interpretarse como la probabilidad aproximada de que la opción al llegar a la fecha de vencimiento de la opción, el precio del activo subyacente esté por encima o debajo del precio de ejercicio dependiendo del tipo de opción. (Saquen, 2020)

El número griego Gamma, por otro lado, mide la tasa de cambio de Delta respecto al precio del activo subyacente. En términos prácticos, una gama alta indica que la delta de la opción es más sensible a cambios en el precio del activo subyacente, lo que significa que la posición puede experimentar cambios significativos en la exposición al riesgo.

Lo que podemos analizar con respecto a las gráficas es que el comportamiento del delta de put y del delta de call, es comportamiento es proporcional entre ambos, aunque signifique cosas completamente contrarias. Lo que podemos observar es que entre mayor certeza acerca del retorno de inversión de la opción call, se tiene una menor certeza de la opción put, cuando ambas tienen el mismo precio de ejercicio. Y por último, podemos observar como Gamma, tiende a mantener valores casi completamente constantes cuando delta no tiene ninguna alteración, pero cuando los valores de delta sufren alguna fluctuación , los valores de gamma llegan a alcanzar sus máximos globales.

```python
# Paso 1: Calcular el precio inicial
P0 = stock_adj['Adj Close'].iloc[-1]

# Paso 2: Simular trayectorias del precio del subyacente
num_simulaciones = 10000
precios_simulados = np.zeros((len(simulated_returns), num_simulaciones))
precios_simulados[0, :] = P0

for t in range(1, len(simulated_returns)):
    precios_simulados[t, :] = precios_simulados[t - 1, :] * np.exp(simulated_returns[t])

# Paso 3: Calcular el valor esperado del subyacente en el futuro
valor_esperado_futuro = np.mean(precios_simulados[-1, :])

print("\nEl valor esperado del subyacente en el futuro es:", valor_esperado_futuro, "\n")

# Plot
plt.plot(precios_simulados)
plt.title('Simulación de Monte Carlo')
plt.xlabel('Día')
plt.ylabel('Precio  ')
plt.show()
```
